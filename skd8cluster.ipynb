```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('like2ben.xls')
data
```

```python
X = data[['hubei','henan','shanxi','jiangxi','hunan','anhui','hebei','fujian']]
```

```python
Xt = X.T
```

# 数据标准化

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(Xt)
data_scaled = pd.DataFrame(data_scaled, columns=data.year)
data_scaled
```

# Ward法进行系统聚类

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.title("理科二本聚类图")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'),labels = ['hubei','henan','shanxi','jiangxi','hunan','anhui','hebei','fujian'])
```
